```python
from pycocotools.coco import COCO
import matplotlib.pyplot as plt
import cv2
import os
import numpy as np
import random

```

```python
cocoRoot = "/mmdetection/mmdetection/data/food_60/"
dataType = ['train', 'val']
annFile_train = os.path.join(cocoRoot, f'annotations/{dataType[0]}.json')
annFile_val = os.path.join(cocoRoot, f'annotations/{dataType[1]}.json')

print(f'Annotation file: {annFile_train}', annFile_val)
```

```python
coco1=COCO(annotation_file=annFile_train)
coco2=COCO(annotation_file=annFile_val)
```
